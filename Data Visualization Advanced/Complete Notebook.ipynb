```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

<h1>Univariate Analysis Code</h1>

This section contains the code for drawing basic plots created for the Univariate Analysis code section in the Tutorial

```python
#Dataset Loading

dataset=pd.read_csv("Datasets/pokemon.csv",index_col=0)

dataset.head(5)

# dataset.tail(5)

#The dataset is very clean(As expected from a Kaggle Dataset)
# pd.isnull(dataset).sum()
```

<h1>Univariate Plots</h1>

```python
dataset.columns
```

```python
#Generating the First bar plot to group pokemon by their Type(Grass,water,fire,fairy,dark etc).Most common type is water

ax=dataset.type1.value_counts().plot.bar(figsize=(12,6),fontsize=15,color="blue")

ax.set_title(label="Pokemon by Primary Type",fontsize=15)

# ax.remove()

#Nice Syntax
plt.box(False)

# sns.despine(bottom=False,left=True)
```

```python
#Most common secondary type is flying.
ax=dataset.type2.value_counts().plot.bar(figsize=(12,6),fontsize=15)

ax.set_title(label="Pokemon by Secondary Type",fontsize=15)

# ax.remove()

#Nice Syntax
# plt.box(False)
```

```python
#The frequency of Pokemon by HP stat total:If we first call sort index,the line chart starts plotting based on values directly

dataset.hp.value_counts().sort_index().plot.line()
```

```python
#Frequency of Pokemon by Weight.Directly call our plotting function as needed

dataset.weight_kg.plot.hist()
```

<h1>Bi variate Analysis</h1>

```python
#Attack vs defense Scatter Plot(Shows coorelation,that pokemon with strong attack have almost an equally strong defense
#Some outliers are present which do not follow this trend)

dataset.plot.scatter(x='attack',y='defense',figsize=(15,6),fontsize=15,title="Attack vs Defense stats for Pokemon")
```

```python
dataset.base_total
```

```python
#The same above plot is not readable so to make it readable we will be using the Hex Plot(Sharex false makes the x axis
#visible)

dataset.plot.hexbin(x='attack',y='defense',gridsize=15,sharex=False)
```

```python
#Check if pokemon are legendary or not generation wise and take their attack and defense mean(#rd Gen legendaries were best)
#For Non legendary pokemon,the stats(Attack,Defense) has remained constant.

pokemon_stats_legendary = dataset.groupby(['is_legendary', 'generation']).mean()[['attack', 'defense']]

pokemon_stats_legendary.plot.bar(stacked=True)
```

```python
#Line Plot To measure basics stats in detail based on generation(Speed has been buffed a few times and sp_attack has increased in the later generations.Attack has also been greatly decreased)

pokemon_stats_by_generation = dataset.groupby('generation').mean()[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]

pokemon_stats_by_generation.plot.line()
```

<h1>Subplotting in Pandas and Matplotlib</h1>

```python
'''Subplots is the process of building multiple plots together and showing it all together.Kind of like building a
workbench to show related plots together'''


figure,subplots=plt.subplots(2,1,figsize=(8,8))

dataset.attack.plot.hist(fontsize=15,color="blue",ax=subplots[0])
subplots[0].set_title("Pokemon Attack Ratings",fontsize=20)

dataset.defense.plot.hist(fontsize=15,color="blue",ax=subplots[1])
subplots[1].set_title("Pokemon Defense Ratings",fontsize=20)

#Added Params to add horizontal space between 2 plots and also changing the font size for the X and Y labels
plt.subplots_adjust(hspace=.4)
plt.rcParams['font.size']=15
```

<h1>Plotting with Seaborn</h1>

```python
'''Seaborn is a better choice tool than matplotlib and pandas built in plotting as it produces slightly better looking 
charts and the syntax is slightly easier'''

#Pokemon count by generation (Gen 5 has the highest pokemon while gen 6 and 7 have the least amount of data)

sns.countplot(dataset.generation)
```

```python
'''Hp variation by pokemon(Lowest to Highest) Max pokemon have around 50 to 100 hp points while very few have greater than
150.The Data seems normally distributed with the data slightly skewed to the right
'''

# sns.distplot(dataset[dataset.hp<=150]['hp'],kde=True)

sns.distplot(dataset.hp)
```

```python
'''Attack vs Defense scatter plot in seaborn(Suggests mild correlation between attack and defense)
Majority of the values are spread between attack 50~75 and defense 40~80
'''

sns.jointplot(dataset.attack,dataset.defense)
```

```python
'''Attack vs Defense Hex plot in seaborn(Suggests mild correlation between attack and defense)
Majority of the values are spread between attack 50~75 and defense 40~80
'''

sns.jointplot(dataset.attack,dataset.defense,kind='hex',gridsize=12)
```

```python
sns.kdeplot(dataset.hp,dataset.attack)
```

```python
'''Using the Box and whiskers plot,we can be certain that legendaries have higher attack overall.The attack distribution 
is almost the same for both legendaries and non legendaries
'''

sns.boxplot(dataset.is_legendary,dataset.attack)
```

```python
sns.violinplot(dataset.is_legendary,dataset.attack)
```

<h1>Facetting</h1>

```python
#Plot to compare attack stats for legendaries and non legendary pokemon

sample_plot=sns.FacetGrid(dataset,row="is_legendary")
sample_plot.map(sns.kdeplot,"attack")
```

```python
#Grouping Attack and Legendaries with generation
sample_plot=sns.FacetGrid(dataset,row="is_legendary",col="generation")
sample_plot.map(sns.kdeplot,"attack")
```

```python
sns.pairplot(dataset[["hp","attack","defense"]])

plt.show()
```

<h1>MultiVariate Plots</h1>

```python
#Multivariate Scallter Plot with Markers and Extra Distinguishability

sns.lmplot(x='attack',y='defense',markers=['x','o'],data=dataset,hue='is_legendary',fit_reg=False)
```

```python
#ViolinPlot to show Total Points across generations

sns.boxplot(x="generation", y="base_total", hue='is_legendary', data=dataset)
```

```python
#Heatmap to show correlation between different variables

sns.heatmap(
    dataset.loc[:, ['hp', 'attack', 'sp_attack', 'defense', 'sp_defense', 'speed']].corr(),
    annot=True
)
```

```python
#Line Chart to Check if we can distinguish fighting pokemon from psychic pokemon(It turns out that we cannot easily distinguish)

from pandas.plotting import parallel_coordinates

dataset2=dataset[dataset.type1.isin(['fighting','psychic'])].loc[:,['type1','attack','sp_attack','defense','sp_defense']]

parallel_coordinates(dataset2,'type1')
```
